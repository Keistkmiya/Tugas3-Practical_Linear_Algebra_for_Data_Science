# Chapter 1: Introduction

## Introduction to Linear Algebra and NumPy

# Chapter 2: Vectors, Part 1

## Vector Algebraic and Geometric Interpretations

Vektor dapat dipahami melalui dua sudut pandang utama: aljabar (algebraic) dan geometri (geometric). Secara aljabar, vektor adalah sebuah daftar baris atau kolom yang berisi angka-angka (skalar) yang merepresentasikan koordinat di dalam ruang multidimensi. Secara geometri, vektor direpresentasikan sebagai sebuah panah yang memiliki arah (direction) dan panjang (magnitude), dimulai dari titik asal atau origin (0,0). NumPy merepresentasikan vektor menggunakan array 1 dimensi untuk operasi matematika standar.

```python
import numpy as np

vector_row = np.array([2, 3])
print(vector_row)
```

## Vector Addition and Subtraction

Operasi penjumlahan dan pengurangan dua vektor dilakukan secara elemen demi elemen (element-wise). Secara geometri, penjumlahan vektor dapat divisualisasikan dengan metode "head-to-tail", di mana pangkal vektor kedua diletakkan di ujung vektor pertama, dan hasil penjumlahannya adalah vektor baru dari titik asal ke ujung vektor kedua. Kedua vektor yang dioperasikan harus memiliki dimensi atau panjang komponen yang sama (homogen).

```python
import numpy as np

v = np.array([4, 1])
w = np.array([2, 3])

v_plus_w = v + w
v_minus_w = v - w

print(v_plus_w)
print(v_minus_w)
```

## Vector-Scalar Multiplication

Perkalian vektor dengan skalar dilakukan dengan mengalikan setiap elemen di dalam vektor dengan nilai skalar tersebut. Secara geometri, operasi ini tidak mengubah arah vektor (kecuali jika skalar bernilai negatif, maka arahnya akan berbalik 180 derajat), melainkan mengubah panjang atau skala dari vektor tersebut (scaling).

```python
import numpy as np

v = np.array([2, 5])
scalar = 3

v_scaled = v * scalar
print(v_scaled)
```

## Vector Dot Product

Dot product (perkalian titik) antara dua vektor menghasilkan sebuah nilai skalar tunggal. Secara aljabar, operasi ini dilakukan dengan mengalikan elemen-elemen yang bersesuaian dari kedua vektor lalu menjumlahkan semua hasil perkalian tersebut. Secara geometri, dot product mencerminkan hubungan sudut antara kedua vektor dan representasi proyeksi satu vektor ke vektor lainnya. Di dalam NumPy, dot product dapat dihitung menggunakan fungsi np.dot() atau operator @.

```python
import numpy as np

v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

dot_product = np.dot(v, w)
print(dot_product)
```

## Vector Length and Unit Vectors

Panjang atau magnitudo dari sebuah vektor (sering disebut sebagai norm atau Euclidean norm) dihitung dengan mencari akar kuadrat dari jumlah kuadrat elemen-elemennya. Unit vector (vektor satuan) adalah vektor yang memiliki panjang tepat bernilai 1. Proses mengubah vektor biasa menjadi vektor satuan disebut normalisasi, yang dilakukan dengan membagi setiap elemen vektor dengan total panjang magnitudonya. Pustaka NumPy menyediakan fungsi np.linalg.norm() untuk menghitung nilai panjang ini.

```python
import numpy as np

v = np.array([3, 4])

v_length = np.linalg.norm(v)
v_unit = v / v_length

print("Panjang vektor:", v_length)
print("Vektor satuan:", v_unit)
print("Panjang vektor satuan:", np.linalg.norm(v_unit))
```

# Chapter 3: Vectors, Part 2

## Vector Independence and Basis

Sekumpulan vektor dikatakan saling bebas linear (linearly independent) jika tidak ada satu pun vektor di dalam kumpulan tersebut yang dapat dinyatakan sebagai kombinasi linear dari vektor-vektor lainnya. Sebaliknya, jika suatu vektor dapat dibentuk dari penjumlahan atau perkalian skalar vektor lain, maka kumpulan tersebut bersifat bergantung linear (linearly dependent). Konsep kebebasan linear ini sangat krusial untuk menentukan basis dari sebuah ruang vektor, di mana basis adalah sekumpulan vektor bebas linear yang dapat menjangkau (span) seluruh ruang tersebut.

```python
import numpy as np

v1 = np.array([1, 2])
v2 = np.array([2, 4])

matrix = np.array([v1, v2])
rank = np.linalg.matrix_rank(matrix)

print("Rank matriks:", rank)
print("Apakah bebas linear?", rank == len(matrix))
```

## Subspaces and Span

Span dari sekumpulan vektor adalah seluruh area atau ruang yang dapat dicapai melalui semua kemungkinan kombinasi linear dari vektor-vektor tersebut. Jika kita memiliki dua vektor di ruang 3D yang tidak searah, maka span dari kedua vektor tersebut akan membentuk sebuah bidang datar 2D (subspace) di dalam ruang 3D. Subspace sendiri adalah ruang vektor kecil yang berada di dalam ruang vektor yang lebih besar, yang tetap memenuhi semua sifat aljabar linear seperti penutupan terhadap operasi penjumlahan dan perkalian skalar.

```python
import numpy as np

v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])

matrix = np.array([v1, v2])
rank = np.linalg.matrix_rank(matrix)

print("Dimensi subspace (Rank):", rank)
```

## Orthogonal and Orthonormal Vectors

Dua buah vektor dikatakan saling ortogonal jika keduanya membentuk sudut tegak lurus (90 derajat). Secara aljabar, kondisi ini terjadi ketika nilai dot product antara kedua vektor sama dengan nol. Jika vektor-vektor yang saling ortogonal tersebut masing-masing memiliki panjang magnitudo tepat bernilai satu (merupakan unit vektor), maka kumpulan vektor tersebut disebut sebagai ortonormal. Basis ortonormal sangat disukai dalam komputasi karena mempermudah proyeksi dan transformasi data tanpa mengubah skala.

```python
import numpy as np

v1 = np.array([1, 0])
v2 = np.array([0, 1])

dot_product = np.dot(v1, v2)
norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)

print("Dot product:", dot_product)
print("Panjang v1:", norm_v1)
print("Panjang v2:", norm_v2)
print("Apakah ortonormal?", dot_product == 0 and norm_v1 == 1 and norm_v2 == 1)
```

# Chapter 4: Vector Applications

## Document Similarity using Cosine Similarity

Salah satu aplikasi utama vektor dalam data science adalah mengukur tingkat kemiripan antar dokumen teks. Dokumen dikonversi menjadi vektor frekuensi kata, lalu sudut antara kedua vektor tersebut dihitung menggunakan formula Cosine Similarity. Nilai Cosine Similarity diperoleh dari hasil pembagian dot product kedua vektor dengan perkalian magnitudo masing-masing vektor. Nilai yang mendekati 1 menunjukkan bahwa kedua dokumen memiliki kemiripan konteks yang tinggi karena arah vektornya hampir searah, sedangkan nilai mendekati 0 menunjukkan tidak ada kemiripan.

```python
import numpy as np

doc1 = np.array([3, 1, 0, 2])
doc2 = np.array([2, 0, 1, 2])

dot_product = np.dot(doc1, doc2)
norm_doc1 = np.linalg.norm(doc1)
norm_doc2 = np.linalg.norm(doc2)

cosine_similarity = dot_product / (norm_doc1 * norm_doc2)
print("Nilai Cosine Similarity:", cosine_similarity)
```

## Vector Projection for Dimension Reduction

Proyeksi vektor (vector projection) adalah proses memetakan sebuah vektor ke atas sumbu atau arah vektor lainnya. Dalam pemrosesan data, teknik ini digunakan untuk dekomposisi data dan reduksi dimensi, di mana komponen informasi yang tidak sejajar dengan sumbu target akan dieliminasi. Secara matematis, proyeksi vektor v pada vektor w dihitung dengan mengalikan hasil pembagian dot product (v dan w) dengan kuadrat magnitudo w, lalu dikalikan kembali dengan vektor w.

```python
import numpy as np

v = np.array([4, 3])
w = np.array([5, 0])

projection = (np.dot(v, w) / np.linalg.norm(w)**2) * w
print("Hasil proyeksi vektor v pada w:", projection)
```
